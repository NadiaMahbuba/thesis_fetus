```python
import tensorflow as tf
from tensorflow.keras import models,layers
import matplotlib.pyplot as plt
```

```python
#abnormal fetus types dataset
IMAGE_SIZE1=224
BATCH_SIZE1=32
base_dir1=r"C:\Abnormal_fetus\train"
dataset1=tf.keras.preprocessing.image_dataset_from_directory(
 base_dir1,
 shuffle=True,
 image_size=(IMAGE_SIZE1,IMAGE_SIZE1),
 batch_size=BATCH_SIZE1   
)
```

```python
class_names1=dataset1.class_names
class_names1
```

```python
plt.figure(figsize=(20,20))
for image_batch1,label_batch1 in dataset1.take(1):
    for i in range(12):
     ax = plt.subplot(3,4,i+1)
     plt.imshow(image_batch1[i].numpy().astype("uint8"))
     plt.title(class_names1[label_batch1[i]])
     plt.axis("off")
```

```python
def get_dataset_partitions_tf1(ds1,train_split1=0.8,val_split1=0.1,test_split1=0.1,shuffle=True,shuffle_size=10000):
    ds_size1=len(ds1)
    if shuffle:
      ds1 = ds1.shuffle(shuffle_size,seed=12)
    train_size1=int(train_split1*ds_size1)
    val_size1=int(val_split1*ds_size1)
    train_ds1=ds1.take(train_size1)
    val_ds1=ds1.skip(train_size1).take(val_size1)
    test_ds1=ds1.skip(train_size1).skip(val_size1)
    return train_ds1,val_ds1,test_ds1  
```

```python
train_ds1,val_ds1,test_ds1=get_dataset_partitions_tf1(dataset1) 
```

```python
train_ds1 = train_ds1.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds1 = val_ds1.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds1 = test_ds1.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
resize_and_rescale1=tf.keras.Sequential([
layers.experimental.preprocessing.Resizing(IMAGE_SIZE1,IMAGE_SIZE1),    
layers.experimental.preprocessing.Rescaling(1.0/255)    
])
```

```python
data_augmentation1=tf.keras.Sequential([
layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
layers.experimental.preprocessing.RandomRotation(0.2)   
])
```

```python
CHANNELS1=3
input_shape1=(BATCH_SIZE1,IMAGE_SIZE1,IMAGE_SIZE1,CHANNELS1)
n_classes1=4
model1=models.Sequential([
resize_and_rescale1,    
layers.Conv2D(filters=64, kernel_size = (3,3), activation="relu", input_shape=input_shape1),
layers.MaxPooling2D((2,2)),
layers.Conv2D(filters=64, kernel_size = (3,3), activation="relu"),
layers.MaxPooling2D((2,2)),
layers.Conv2D(filters=128, kernel_size = (3,3), activation="relu"),
layers.MaxPooling2D((2,2)),
layers.Conv2D(filters=128, kernel_size = (3,3), activation="relu"),
layers.MaxPooling2D((2,2)),
layers.Conv2D(filters=256, kernel_size = (3,3), activation="relu"),
layers.MaxPooling2D((2,2)),        
layers.Flatten(),
layers.Dense(64,activation='relu'),
layers.Dense(128,activation='relu'),     
layers.Dense(n_classes1,activation='softmax'),
])
```

```python
model1.build(input_shape=input_shape1)
```

```python
model1.summary()
```

```python
#first model for abnormal fetus dataset
model1.compile(
optimizer='adam',
loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
metrics=['accuracy']    
)
```

```python
history1=model1.fit(
train_ds1,
epochs=50,
batch_size=BATCH_SIZE1,
verbose=1,
validation_data=val_ds1    
)
```

```python
model1.evaluate(test_ds1)
```

```python
import tensorflow as tf
from tensorflow.keras import models,layers
import matplotlib.pyplot as plt
IMAGE_SIZE=256
BATCH_SIZE=32
base_dir=r"C:\fetus_practice\train"
dataset=tf.keras.preprocessing.image_dataset_from_directory(
 base_dir,
 shuffle=True,
 image_size=(IMAGE_SIZE,IMAGE_SIZE),
 batch_size=BATCH_SIZE   
)
```

```python
class_names=dataset.class_names
class_names
```

```python
plt.figure(figsize=(20,20))
for image_batch,label_batch in dataset.take(1):
    for i in range(12):
     ax = plt.subplot(3,4,i+1)
     plt.imshow(image_batch[i].numpy().astype("uint8"))
     plt.title(class_names[label_batch[i]])
     plt.axis("off")
```

```python
def get_dataset_partitions_tf(ds,train_split=0.8,val_split=0.1,test_split=0.1,shuffle=True,shuffle_size=10000):
    ds_size=len(ds)
    if shuffle:
      ds = ds.shuffle(shuffle_size,seed=12)
    train_size=int(train_split*ds_size)
    val_size=int(val_split*ds_size)
    train_ds=ds.take(train_size)
    val_ds=ds.skip(train_size).take(val_size)
    test_ds=ds.skip(train_size).skip(val_size)
    return train_ds,val_ds,test_ds  
```

```python
train_ds,val_ds,test_ds=get_dataset_partitions_tf(dataset) 
```

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
resize_and_rescale=tf.keras.Sequential([
layers.experimental.preprocessing.Resizing(IMAGE_SIZE,IMAGE_SIZE),    
layers.experimental.preprocessing.Rescaling(1.0/255)    
])
```

```python
data_augmentation=tf.keras.Sequential([
layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
layers.experimental.preprocessing.RandomRotation(0.2)   
])
```

```python
CHANNELS=3
input_shape=(BATCH_SIZE,IMAGE_SIZE,IMAGE_SIZE,CHANNELS)
n_classes=2
model=models.Sequential([
resize_and_rescale,    
layers.Conv2D(filters=64, kernel_size = (3,3), activation="relu", input_shape=input_shape),
    layers.MaxPooling2D((2,2)),
layers.Conv2D(filters=64, kernel_size = (3,3), activation="relu"),
layers.MaxPooling2D((2,2)),
layers.Conv2D(filters=128, kernel_size = (3,3), activation="relu"),
layers.MaxPooling2D((2,2)),
layers.Conv2D(filters=256, kernel_size = (3,3), activation="relu"),
layers.MaxPooling2D((2,2)),
layers.Conv2D(filters=256, kernel_size = (3,3), activation="relu"),
layers.MaxPooling2D((2,2)),        
layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(128,activation='relu'),
     layers.Dense(n_classes,activation='sigmoid'),
])
```

```python
model.build(input_shape=input_shape)
```

```python
model.compile(
optimizer='adam',
loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
metrics=['accuracy']    
)
```

```python
history=model.fit(
train_ds,
epochs=50,
batch_size=BATCH_SIZE,
verbose=1,
validation_data=val_ds   
)
```

```python
model.evaluate(test_ds)
```

```python
#detecting healthy and abnormal and denote abnormal fetus typeimport numpy as np
def predict(model,img):
    img_array=tf.keras.preprocessing.image.img_to_array(img)
    img_array=tf.expand_dims(img_array,0)
    predictions=model.predict(img_array)
    predicted_class=class_names[np.argmax(predictions[0])]
    confidence=round(100 * (np.max(predictions[0])) , 2)
    return predicted_class,confidence #predict function return predicted_class,confidence
def abnormal_class_type(model,img):
    img_array=tf.keras.preprocessing.image.img_to_array(img)
    img_array=tf.expand_dims(img_array,0)
    predictions=model.predict(img_array)
    abnormal_type=class_names1[np.argmax(predictions[0])]
    return abnormal_type
```

```python
plt.figure(figsize=(15,15))
for images,labels in test_ds.take(1):
    for i in range(9):
        ax=plt.subplot(3,3,i+1)
        plt.imshow(images[i].numpy().astype("uint8"))
        predicted_class,confidence=predict(model,images[i].numpy())
        actual_class=class_names[labels[i]]
        plt.title(f"Actual: {actual_class},\nPredicted:{predicted_class}.\nAccuracy:{confidence}%")
        if predicted_class == "Abnormal":
            abnormal_type=abnormal_class_type(model1,images[i].numpy())
            plt.title(f"Actual: {actual_class},\nPredicted:{predicted_class}.\n Accuracy:{confidence}% \n Abnormal type:{abnormal_type}")
        plt.axis("off")
```
